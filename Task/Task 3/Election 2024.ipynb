```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

final_result_list = []

for j in range(11, 2526):
    url = 'https://results.eci.gov.in/PcResultGenJune2024/candidateswise-S{:04d}.htm'.format(j)
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.0.0 Safari/537.36'}
    webpage = requests.get(url, headers=headers).text
    soup = BeautifulSoup(webpage, 'lxml')
    
    state = []
    assembly = []
    winner = []
    winner_party = []
    winner_votes = []
    runner_up = []
    margin = []
    runner_up_party = []
    runner_up_votes = []
    
    # Extract state name
    state_name = None
    state_strong = soup.find_all('strong')
    for i in state_strong:
        text = i.text
        start = text.find('(') + 1
        end = text.find(')')
        extract = text[start:end]
        if extract:
            state_name = extract.strip()
            
    # Extract Assembly Constituency
    for i in soup.find_all('h2'):
        sp = i.find('span')
        if sp:
            text = i.text
            start = text.find('-') + 1
            end = text.find('(') - 1
            extract = text[start:end]
            assembly.append(extract.strip())
    
    # Extract candidate details
    AP = soup.find_all('div', class_='box-wraper box-boarder')
    for i in AP:
        name = i.find('div', class_='cand-box')
        name2 = i.find_all('div', 'cand-box')
        
        if name:
            final = name.find('div', class_='nme-prty')
            status_won = name.find('div', class_='status won')
            
            if final:
                m = final.find('h5')
                n = final.find('h6')
                if m:
                    winner.append(m.text.strip())
                else:
                    winner.append(None)
                if n:
                    winner_party.append(n.text.strip())
                else:
                    winner_party.append(None)
            
            if status_won:
                spans = status_won.find('span')
                
                    
#                 for span in spans:
                text = status_won.text
                start = text.find('won') + 4
                end = text.find('(')
                extract = text[start:end]
                if len(extract) > 0:
                    winner_votes.append(extract.strip())
                else:
                    winner_votes.append(None)
                if spans:
                    text2 = spans.text
                    start2 = text2.find('+') + 1
                    end2 = text2.find(')')
                    extract2 = text2[start2:end2]
                
                    if len(extract2) > 0:
                    
                        margin.append(extract2.strip())
                    else:
                        margin.append(None)
            else:
                winner_votes.append(None)
                margin.append(None)
            
        if len(name2) > 1:
            final2 = name2[1].find('div', class_='nme-prty')
            status_lost = name2[1].find('div', class_='status lost')

            if final2:
                mn = final2.find('h6')
                mn2 =final2.find('h5')
                if mn:
                    runner_up_party.append(mn.text.strip())
                    runner_up.append(mn2.text.strip())
                else:
                    runner_up_party.append(None)
            else:
                runner_up_party.append(None)
                
            if status_lost:
                text = status_lost.text
                start = text.find('lost')+4
                end = text.find('(')
                extract = text[start:end]
#                 spans = status_lost.find_all('span')

                if len(extract) > 0:
                    runner_up_votes.append(extract.strip())
                else:
                    runner_up_votes.append(None)
            else:
                runner_up_votes.append(None)

    num_candidates = len(winner)
    if state_name:
        state = [state_name] * num_candidates
    else:
        state = [None] * num_candidates
    
    data = pd.DataFrame({
        'State': state,
        'Assembly Constituency': assembly,
        'Winner': winner,
        'Winner Party': winner_party,
        'Winner Votes': winner_votes,
        'Runner-up':runner_up,
        'Runner-up Party': runner_up_party,
        'Runner-up Votes': runner_up_votes,
        'Margin': margin
    })
    
    final_result_list.append(data)

final_result = pd.concat(final_result_list, ignore_index=True)
print(final_result)
```

```python

import requests
from bs4 import BeautifulSoup
import pandas as pd

final_result_list = []

for j in range(11, 2527):  # Adjust range as needed, here 121 for testing
    url = 'https://results.eci.gov.in/PcResultGenJune2024/candidateswise-S{:03d}.htm'.format(j)
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.0.0 Safari/537.36'}
    webpage = requests.get(url, headers=headers).text
    soup = BeautifulSoup(webpage, 'lxml')
    
    state = []
    assembly = []
    winner = []
    winner_party = []
    winner_votes = []
    runner_up = []
    margin = []
    runner_up_party = []
    runner_up_votes = []
    
    # Extract state name
    state_name = None
    state_strong = soup.find_all('strong')
    for i in state_strong:
        text = i.text
        start = text.find('(') + 1
        end = text.find(')')
        extract = text[start:end]
        if extract:
            state_name = extract.strip()
            
    # Extract Assembly Constituency
    for i in soup.find_all('h2'):
        sp = i.find('span')
        if sp:
            text = i.text
            start = text.find('-') + 1
            end = text.find('(') - 1
            extract = text[start:end]
            assembly.append(extract.strip())
    
    # Extract candidate details
    AP = soup.find_all('div', class_='box-wraper box-boarder')
    for i in AP:
        name = i.find('div', class_='cand-box')
        name2 = i.find_all('div', class_='cand-box')
        
        if name:
            final = name.find('div', class_='nme-prty')
            status_won = name.find('div', class_='status won')
            
            if final:
                m = final.find('h5')
                n = final.find('h6')
                if m:
                    winner.append(m.text.strip())
                else:
                    winner.append(None)
                if n:
                    winner_party.append(n.text.strip())
                else:
                    winner_party.append(None)
            
            if status_won:
                spans = status_won.find_all('span')
                text = status_won.text
                start = text.find('won') + 4
                end = text.find('(')
                extract = text[start:end].strip()
                winner_votes.append(extract if extract else None)
                
                if spans:
                    text2 = spans[0].text
                    start2 = text2.find('+') + 1
                    end2 = text2.find(')')
                    extract2 = text2[start2:end2].strip()
                    margin.append(extract2 if extract2 else None)
                else:
                    margin.append(None)
            else:
                winner_votes.append(None)
                margin.append(None)
        
        if len(name2) > 1:
            final2 = name2[1].find('div', class_='nme-prty')
            status_lost = name2[1].find('div', class_='status lost')

            if final2:
                mn = final2.find('h6')
                mn2 = final2.find('h5')
                if mn:
                    runner_up_party.append(mn.text.strip())
                    runner_up.append(mn2.text.strip())
                else:
                    runner_up_party.append(None)
                    runner_up.append(None)
            else:
                runner_up_party.append(None)
                runner_up.append(None)
                
            if status_lost:
                text = status_lost.text
                start = text.find('lost') + 5
                end = text.find('(')
                extract = text[start:end].strip()
                runner_up_votes.append(extract if extract else None)
            else:
                runner_up_votes.append(None)

    num_candidates = len(winner)
    if state_name:
        state = [state_name] * num_candidates
    else:
        state = [None] * num_candidates
    
    data = pd.DataFrame({
        'State': state,
        'Assembly Constituency': assembly,
        'Winner': winner,
        'Winner Party': winner_party,
        'Winner Votes': winner_votes,
        'Runner-up': runner_up,
        'Runner-up Party': runner_up_party,
        'Runner-up Votes': runner_up_votes,
        'Margin': margin
    })
    
    final_result_list.append(data)

final_result = pd.concat(final_result_list, ignore_index=True)
print(final_result)
```
