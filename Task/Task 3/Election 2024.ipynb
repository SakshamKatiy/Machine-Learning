```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Initialize an empty list to store DataFrames
final_result_list = []

# Loop through the URLs
for j in range(11, 50):
    url = 'https://results.eci.gov.in/PcResultGenJune2024/candidateswise-S0{}.htm'.format(j)
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.0.0 Safari/537.36'}
    webpage = requests.get(url, headers=headers).text
    soup = BeautifulSoup(webpage, 'lxml')
    
    # Initialize lists to store the data
    state = []
    winner = []
    winner_party = []
    winner_votes = []
    margin = []
    runner_up_party = []
    runner_up_votes = []
    
    # Extract state name (assuming it's constant for the entire page)
    state_name = None
    state_strong = soup.find_all('strong')
    for i in state_strong:
        text = i.text
        start = text.find('(')+1
        end = text.find(')') 
        extract = text[start:end]
        if extract:
            state_name = extract.strip()
    
    # Find all the candidate boxes
    AP = soup.find_all('div', class_='box-wraper box-boarder')
    for i in AP:
        name = i.find('div', class_='cand-box')
        if name:
            # Extract winner and runner-up details
            final = name.find('div', class_='nme-prty')
            status_won = name.find('div', class_='status won')
            status_lost = name.find('div', class_='status lost')
            
            if final:
                m = final.find('h5')
                n = final.find('h6')
                if m:
                    winner.append(m.text.strip())
                else:
                    winner.append(None)
                if n:
                    winner_party.append(n.text.strip())
                    runner_up_party.append(n.text.strip())
                else:
                    winner_party.append(None)
            else:
                winner.append(None)
                winner_party.append(None)
            
            if status_won:
                spans = status_won.find_all('span')
                for i in spans:
                    text = i.text
                    start = text.find('+')+1
                    end = text.find(')')
                    extract = text[start:end]
                if len(extract) > 0:
                    winner_votes.append(extract.strip())
                else:
                    winner_votes.append(None)
                if len(spans) > 1:
                    margin.append(spans[1].text.strip())
                else:
                    margin.append(None)
            else:
                winner_votes.append(None)
                margin.append(None)
            
            if status_lost:
                runner_up_votes.append(status_lost.text.strip())
                spans = status_lost.find_all('span')
                if len(spans) > 0:
                    runner_up_votes.append(spans[0].text.strip())
                else:
                    runner_up_votes.append(None)
                if len(spans) > 1:
                    runner_up_party.append(spans[1].text.strip())
                else:
                    runner_up_party.append(None)
            else:
                runner_up_votes.append(None)
#                 runner_up_party.append(None)
    
    # Ensure all lists have the same length by adjusting state list
    num_candidates = len(winner)
    if state_name:
        state = [state_name] * num_candidates
    else:
        state = [None] * num_candidates
    
    # Create a DataFrame for the current page
    data = pd.DataFrame({
        'State': state,
        'Winner': winner,
        'Winner Party': winner_party,
        'Winner Votes': winner_votes,
        'Runner-up Party': runner_up_party,
        'Runner-up Votes': runner_up_votes,
        'Margin': margin
    })
    
    # Append the DataFrame to the list
    final_result_list.append(data)

# Concatenate all DataFrames in the list
final_result = pd.concat(final_result_list, ignore_index=True)

# Display the final result
print(final_result)
```
