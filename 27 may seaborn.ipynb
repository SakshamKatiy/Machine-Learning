```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.get_dataset_names()
```

```python
data
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.sample(5)
```

```python
x = tips['total_bill']
y = tips['tip']
```

```python
sns.scatterplot(x='total_bill',y = 'tip',data = tips,hue ='time')
```

```python
sns.relplot(data=tips,x='total_bill',y = 'tip',hue = 'time',sizes = (10,1000),palette='Accent_r')
plt.xlabel('Total bill in USD $')
plt.ylabel('Tips to the service')
plt.xlabel('Total bill in USD $')
plt.title('New York')
```

```python
sns.lineplot(data = tips,x = 'total_bill',y='size',hue='sex',markers=True,style='smoker',dashes = True)
```

```python
sns.barplot(y='total_bill',x = 'size',data=tips)
```

```python
sns.histplot(x='size',data=tips,hue='sex',bins= 20)
```

```python
sns.histplot(x='size',data=tips,kde =True, bins=5,hue='smoker',palette='Accent')
```

```python
sns.kdeplot(x ='size',data=tips,hue='smoker',fill = True)
```
